```python
# find the number of movies starting with A for each group

def foo(group):
     return group['Series_Title'].str.startswith('A').sum()
```

```python
genres.apply(foo)
```

```python
# find the ranking of each movie in the group according to IRAM rank
def rank_movie(group):
     group['genre_rank'] = group['IMDB_Rating'].rank(ascending= False)
     return group
```

```python
genres.apply(rank_movie)
```

```python
# find normalized IMDB rating group -> normaliztion rank -> x-x(min)/x(max)-x(min)


def normal(group):
    group["normalizarion"] = (
        group["IMDB_Rating"]
        - group["IMDB_Rating"].min() / group["IMDB_Rating"].max()
        - group["IMDB_Rating"].min()
    )

    return group


genres.apply(normal)
```
